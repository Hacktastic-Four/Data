```python
import numpy as np
import pandas as pd
```

```python
django = pd.read_csv('django_data.csv')
flutter = pd.read_csv('flutter_data.csv')
ml = pd.read_csv('ml_data.csv')
react = pd.read_csv('react_data.csv')
solidity = pd.read_csv('solidity_data.csv')
```

```python
django.head()
```

```python
ls_header = []
ls_full_ques = []
ls_link = []
ls_answers = []
ls_category = []
```

```python
for i in range(len(django)):
    ls_header.append(django["Header-Question"][i])
    if(len(str(django["Full-Question"][i]))>=3):
        ls_full_ques.append(str(django["Full-Question"][i])[2:])
    else:
        ls_full_ques.append(django["Full-Question"][i])
    ls_link.append(django["Links"][i])
    if(len(str(django["Answers"][i]))>=3):
        ls_answers.append(str(django["Answers"][i])[2:])
    else:
        ls_answers.append(django["Answers"][i])
    ls_category.append("django")
```

```python
for i in range(len(react)):
    ls_header.append(react["Header-Question"][i])
    if(len(str(react["Full-Question"][i]))>=3):
        ls_full_ques.append(str(react["Full-Question"][i])[2:])
    else:
        ls_full_ques.append(react["Full-Question"][i])
    ls_link.append(react["Links"][i])
    if(len(str(react["Answers"][i]))>=3):
        ls_answers.append(str(react["Answers"][i])[2:])
    else:
        ls_answers.append(react["Answers"][i])
    ls_category.append("react")
```

```python
for i in range(len(flutter)):
    ls_header.append(flutter["Header-Question"][i])
    if(len(str(flutter["Full-Question"][i]))>=3):
        ls_full_ques.append(str(flutter["Full-Question"][i])[2:])
    else:
        ls_full_ques.append(flutter["Full-Question"][i])
    ls_link.append(flutter["Links"][i])
    if(len(str(flutter["Answers"][i]))>=3):
        ls_answers.append(str(flutter["Answers"][i])[2:])
    else:
        ls_answers.append(flutter["Answers"][i])
    ls_category.append("flutter")
```

```python
for i in range(len(solidity)):
    ls_header.append(solidity["Header-Question"][i])
    if(len(str(solidity["Full-Question"][i]))>=3):
        ls_full_ques.append(str(solidity["Full-Question"][i])[2:])
    else:
        ls_full_ques.append(solidity["Full-Question"][i])
    ls_link.append(solidity["Links"][i])
    if(len(str(solidity["Answers"][i]))>=3):
        ls_answers.append(str(solidity["Answers"][i])[2:])
    else:
        ls_answers.append(solidity["Answers"][i])
    ls_category.append("solidity")
```

```python
for i in range(len(ml)):
    ls_header.append(ml["Header-Question"][i])
    if(len(str(ml["Full-Question"][i]))>=3):
        ls_full_ques.append(str(ml["Full-Question"][i])[2:])
    else:
        ls_full_ques.append(ml["Full-Question"][i])
    ls_link.append(ml["Links"][i])
    if(len(str(ml["Answers"][i]))>=3):
        ls_answers.append(str(ml["Answers"][i])[2:])
    else:
        ls_answers.append(ml["Answers"][i])
    ls_category.append("ml")
```

# Final List

```python
super_ls = []
for i in range(len(ls_link)):
    super_ls.append([ls_link[i], ls_header[i], ls_category[i], ls_full_ques[i], ls_answers[i]])
```

```python
dataframe  = pd.DataFrame(super_ls, columns=['Links', 'Header-Question', 'Category','Full-Question', 'Answers'])
```
